El tipo de gráfico más básico que puede hacer es un gráfico de línea con datos definidos como listas de Python o matrices numpy.Como veremos, los parámetros pueden ser cualquier valor de matriz o escalar.

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 3, 4, 5, 6, 7]

plt.plot(x, y);
```

Esto nos da un gráfico de línea muy simple con escamas del eje X e Y pero sin etiquetas.El tamaño y el color de la línea de trama usan los valores predeterminados de Matplotlib.Cambie el color estableciendo un atributo `color`.

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([1, 1, 2, 3, 5, 8, 13])

plt.plot(x, y, color='red');
```

<b> Marcadores </b>
</div>

También puede controlar los marcadores utilizados para trazar cada punto de datos.

```python
y = [2, 5, 9, 13, 18]

plt.plot(y, marker='o');  # That's the lowercase letter 'o'
```

Hay muchos marcadores diferentes para elegir.

```python
y1 = [1, 2, 3, 4, 5]
y2 = [2, 5, 9, 13, 18]

plt.plot(y1, marker='s')  # square marker
plt.plot(y2, marker='D'); # diamond marker
```

También puede cambiar el tamaño del marcador.

```python
y = [2, 5, 9, 13, 18]

plt.plot(y, marker='o', markersize=12);
```

También podemos cambiar el color del borde del marcador y el color de la cara.

```python
plt.plot(y, marker='o', markersize=12, markeredgecolor='red', markerfacecolor='yellow');
```

Puede ver todos los diferentes estilos de marcadores utilizando `matplotlib.markers.markerstyle.markers`, o ver todos los símbolos disponibles en [matplotlib.markers] (https://matplotlib.org/stable/api/markers_api.html).

Dado que Matplotlib toma cualquier valor de matriz o escalar como parámetros de datos, podemos usarlo para trazar funciones continuas (si solo le damos suficientes puntos de datos para trazar).

```python
x = np.linspace(start=-np.pi, stop=np.pi, num=256)
y = np.sin(2*x)

plt.plot(x, y);
```

También podemos llenar los colores entre dos líneas en la misma trama.En el cuadro a continuación, observe que el segundo conjunto de valores Y es un escalar.Esto le dice a Matplotlib que llene el área entre la primera función y la coordenada 0 Y.

Si observa la gráfica anterior, puede ser difícil estimar los valores de los puntos de datos en cada línea de trama.Agreguemos marcadores y líneas de cuadrícula para facilitar esto.Para este primer ejemplo, solo llamaré a `Grid (verdadero)` para mostrar líneas de cuadrícula con la configuración predeterminada.

```python
x = ['Jan', 'Feb', 'March', 'Apr', 'May']
y1 = [1, 2, 3, 4, 5]
y2 = [2, 5, 9, 13, 18]
y3 = [2, 4, 6, 8, 10]

plt.plot(x, y1, x, y2, x, y3, marker='o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend(labels=['Alice', 'Bob', 'Carol'], loc='upper left')
plt.grid(True);
```

Consulte la documentación en [`matplotlib.pyplot.grid`] (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html) para ver todos los parámetros opcionales que puede pasar para controlar elestilo de las líneas de la cuadrícula.Solo para tener una idea de lo que es posible, cambiemos el color y el grosor de la línea, pero también el alfa para hacer que las líneas de cuadrícula sean más transparentes.

```python
x = ['Jan', 'Feb', 'March', 'Apr', 'May']
y1 = [1, 2, 3, 4, 5]
y2 = [2, 5, 9, 13, 18]
y3 = [2, 4, 6, 8, 10]

plt.plot(x, y1, x, y2, x, y3, marker='o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend(labels=['Alice', 'Bob', 'Carol'], loc='upper left')
plt.grid(color='blue', alpha=0.1, linewidth=3);
```

Los datos para una gráfica de barra son diferentes de los datos que hemos visto hasta ahora.Para los gráficos de línea, estábamos usando datos numéricos en los ejes X e Y.Para un gráfico de barras, necesitamos que nuestros datos del eje X sean un conjunto de etiquetas para las barras, y los datos del eje Y deben ser un valor numérico que corresponde a la altura de cada barra.

Creemos datos para nuestro primer gráfico de barras utilizando la columna 'Embarked` del conjunto de datos Titanic.

```python
embarked = titanic_df['Embarked'].value_counts()
print(embarked.index) # these will be our bar labels
print(embarked.values) # these will be the heights of the bars
```

Creamos una gráfica de barra usando [`matplotlib.pyplot.bar`] (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html).

```python
plt.bar(x=embarked.index, height=embarked.values);
```

También podemos especificar una lista de colores para aplicar a cada barra.

```python
plt.bar(x=embarked.index, height=embarked.values, color=['blue', 'green', 'red']);
```

También podemos crear una gráfica de barra horizontal utilizando la función 'Barh`, pero tenemos que especificar las etiquetas en el eje Y los valores como anchos de barra en lugar de alturas.

```python
plt.barh(y=embarked.index, width=embarked.values, color=['blue', 'green', 'red']);
```

Para crear gráficos de barras apiladas, podemos separar los datos embarcados en dos grupos para hombres y mujeres.

```python
embarked_m = titanic_df[titanic_df['Sex']=='male']['Embarked'].value_counts()
embarked_f = titanic_df[titanic_df['Sex']=='female']['Embarked'].value_counts()
print(embarked_m)
print(embarked_f)
```

Para apilar las barras, trazamos un conjunto de datos normalmente, luego le decimos a Matplotlib que apilara al otro en la parte superior utilizando el parámetro 'Bottom`.Esto especifica la coordenada Y de las bases de las barras.

```python
plt.bar(x=embarked_m.index, height=embarked_m.values, color='blue')
plt.bar(x=embarked_f.index, height=embarked_f.values, bottom=embarked_m.values, color='red');
```

Finalmente, podemos usar las mismas funciones que vimos antes para proporcionar etiquetas del eje, un título y una leyenda de la trama.

```python
plt.bar(x=embarked_m.index, height=embarked_m.values, color='blue', width=0.75)
plt.bar(x=embarked_f.index, height=embarked_f.values, bottom=embarked_m.values, color='red', width=0.75)

plt.xlabel('Port')
plt.ylabel('Passengers')
plt.title('Port of Embarkation by Sex')
plt.legend(labels=['M', 'F']);
```

Podemos usar Matplotlib's [Hist`] (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) Función para mostrar una distribución de frecuencia de alguna variable.Mostremos la distribución de las edades de los pasajeros en el Titanic.

```python
ages = titanic_df['Age']
plt.hist(ages);
```

Tenga en cuenta que esto ilustra el recuento total (eje y) de las edades (eje x) de los pasajeros separados en 10 contenedores.Podemos mostrar una densidad de probabilidad de edades en cada contenedor en lugar del recuento total cambiando a una tabla de densidad.

```python
plt.hist(ages, density=True);
```

También podemos cambiar el número de contenedores del valor predeterminado para que el histograma sea más o menos granular, y agregar títulos, etiquetas y líneas de cuadrícula para que sea más legible.

```python
plt.hist(ages, bins=20)

plt.xlabel('Age')
plt.ylabel('Passengers')
plt.title('Distribution of Passengers Ages')
plt.grid(True);
```

A partir de esta distribución, podemos ver que la mayoría de los pasajeros estaban en el rango de edad de aproximadamente 20 a 40. También podemos ver que había muchos niños pequeños a bordo y relativamente pocos pasajeros mayores de 70 años.

Anteriormente mostramos cómo el color y el tamaño del marcador pueden usarse como una tercera dimensión en matplotlib, pero también puede trazar esos datos en un tercer eje estableciendo la proyección a `3D` al llamar a la función 'ejes'.Esa función devuelve un objeto de eje que usamos para crear un diagrama de dispersión.

```python
age = titanic_df['Age']
fare = titanic_df['Fare']
pclass = titanic_df['Pclass']

fig = plt.figure(figsize=(9, 9))

ax = plt.axes(projection='3d')
ax.scatter(xs=age, ys=fare, zs=pclass)

ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Class');
```

```python
fig = plt.figure(figsize=(9, 9))

ax = plt.axes(projection='3d')
ax.scatter(xs=age, ys=fare, zs=pclass, c=pclass, cmap='brg')

ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Class');
```

Muchas veces necesitarás trazar múltiples cuadros en una figura.Matplotlib le permite hacer esto con la función [`matplotlib.pyplot.subplots`] (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html), donde se especifica el número de ROWcolumnas, y devuelve una cuadrícula de ejes.Esto tendrá más sentido con un ejemplo.Dibujemos dos gráficos circulares del conjunto de datos Titanic en una cuadrícula, uno al lado del otro.

```python
s_vals = titanic_df['Sex'].value_counts()
e_vals = titanic_df['Embarked'].value_counts()

fig, axes = plt.subplots(1, 2) # create a grid with 1 row, 2 columns
fig.set_size_inches(8, 4) # make the figure wide

# Crea un gráfico circular en la primera ranura de la cuadrícula
axes[0].pie(s_vals, labels=['Male', 'Female'], startangle=90, autopct='%.1f%%')
axes[0].set_title('Sex')

# Crea un gráfico circular en la segunda ranura de la cuadrícula
axes[1].pie(e_vals, labels=['S', 'C', 'Q'], startangle=90, explode=[0, 0.1, 0])
axes[1].set_title('Embarked');
```

No tiene que crear el mismo tipo de gráfico en cada espacio de la cuadrícula.

```python
fig, axes = plt.subplots(1, 3) # create a grid with 1 row, 3 columns
fig.set_size_inches(12, 4) # make the figure wide

# Crea un gráfico circular en la primera ranura de la cuadrícula
axes[0].pie(s_vals, labels=['Male', 'Female'], startangle=90, autopct='%.1f%%')
axes[0].set_title('Sex')

# Crea una trama de caja en la segunda ranura de la cuadrícula
age = titanic_df['Age'].dropna() # Drop missing values before plotting or the boxplot will be empty
axes[1].boxplot(age, labels=['Age']);

# Crear una trama de dispersión en la tercera ranura de la cuadrícula
age = titanic_df['Age']
fare = titanic_df['Fare']
axes[2].scatter(x=age, y=fare);
```

<b> Texto y anotaciones </b>

Matplotlib le permite agregar texto y anotaciones a una trama para que pueda etiquetar las características de sus datos a los que le gustaría llamar la atención.Volvamos a nuestro ejemplo anterior de trazar una función, pero esta vez agregaremos una etiqueta que dice cuál es la función.

```python
x = np.linspace(start=-np.pi, stop=np.pi, num=256)
y = np.sin(2*x)

fig = plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.text(-1.65, 0.5, 'y = sin(2x)', fontsize=14);
```

También podríamos trazar y anotar un punto específico.

```python
fig = plt.figure(figsize=(8, 5))
plt.plot(x, y, color='gray')

plt.plot([0], [0], 'o') # plot the point (0, 0)

arrowprops = dict(facecolor='black', shrink=0.1, width=2)
plt.annotate('origin point',        # text of the annotation
             xy=(0, 0),             # where the arrow should point
             xytext=(-1.65, 0.5),   # Where the text should start
             arrowprops=arrowprops, # arrow properties
             fontsize=12            # text font size
            );
```
